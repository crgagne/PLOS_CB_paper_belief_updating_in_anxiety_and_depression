### Figure 4 (Mediation Analyses)

```python
traits = ['item_clinical.fa.omega3.g',
        'item_clinical.fa.omega3.anh',
        'item_clinical.fa.omega3.cog_anx']
data = df[['%_selected_everyone','MID']+traits]
data = data.rename(columns={'%_selected_everyone':'selected'})
data =data.rename(columns={ 'item_clinical.fa.omega3.anh':'andep',
                             'item_clinical.fa.omega3.g':'g',
                             'item_clinical.fa.omega3.cog_anx':'coganx'}) #{}
data = pd.merge(data,model_fits_df[['u_u0','MID']],on='MID')
data['belief_accuracy'] = np.sqrt((data['selected']-data['u_u0'])**2)
data.head()
```

```python
data = sm.add_constant(data)
data.head()
```

```python
import statsmodels.api as sm
from statsmodels.stats.mediation import Mediation

outcome_model = sm.OLS.from_formula("u_u0~ andep + selected + const ",data)
mediator_model = sm.OLS.from_formula("selected ~andep + const", data)
```

```python
med = Mediation(outcome_model, mediator_model, "andep", "selected").fit()
```

```python
med.summary()
```

```python
mediator_model.fit().summary()
```

```python
basic_model = sm.OLS.from_formula("u_u0 ~ andep", data)
basic_model.fit().summary()
```

#### Partialed model 

```python
resid_model = sm.OLS.from_formula("andep ~ selected", data)
data['andep_wo_selected']= resid_model.fit().resid.values

partialled_model = sm.OLS.from_formula("u_u0 ~ andep_wo_selected", data)
partialled_model.fit().summary()
```

```python
resid_model_2 = sm.OLS.from_formula("u_u0 ~ selected", data)
data['u_u0_wo_selected']= resid_model_2.fit().resid.values #+ resid_model_2.fit().params.const

resid_model_2.fit().summary()
```
